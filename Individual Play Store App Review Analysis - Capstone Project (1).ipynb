#**DATA MANIPULATION**

```python
def assign_median(series):
  return series.fillna(series.median)
```

```python
df3.Rating= df3['Rating'].transform(assign_median)
```

```python
print(df3['Type'].mode())
print(df3['Current Ver'].mode())
print(df3['Android Ver'].mode())
print(df3['Content Rating'].mode())
print(df3['Genres'].mode())
```

```python
df3['Type'].fillna(str(df3['Type'].mode().values[0]), inplace = True)
df3['Current Ver'].fillna(str(df3['Current Ver'].mode().values[0]), inplace = True)
df3['Android Ver'].fillna(str(df3['Android Ver'].mode().values[0]), inplace = True)
df3['Category'].fillna(str(df3['Category'].mode().values[0]), inplace = True)
df3['Reviews'].fillna(str(df3['Reviews'].mode().values[0]), inplace = True)
df3['Size'].fillna(str(df3['Size'].mode().values[0]), inplace = True)
df3['Installs'].fillna(str(df3['Installs'].mode().values[0]), inplace = True)
df3['Price'].fillna(str(df3['Price'].mode().values[0]), inplace = True)
df3['Last Updated'].fillna(str(df3['Last Updated'].mode().values[0]), inplace = True)
df3['Translated_Review'].fillna(str(df3['Translated_Review'].mode().values[0]), inplace = True)
df3['Sentiment'].fillna(str(df3['Sentiment'].mode().values[0]), inplace = True)
df3['Sentiment_Polarity'].fillna(str(df3['Sentiment_Polarity'].mode().values[0]), inplace = True)
df3['Sentiment_Subjectivity'].fillna(str(df3['Sentiment_Subjectivity'].mode().values[0]), inplace = True)

```

#**FIXING NUMERICAL DATA**



```python
for review in df3['Reviews'].unique():
  print(review)
```

```python
df3['Reviews'] = pd.to_numeric(df3['Reviews'], errors='coerce')
df3
```

```python
df3.info()
```

```python
for s in df3['Size'].unique():
  print(s)
```

```python
def fix_Size(s):
    try:
        if s.endswith('M'):
             return float(s[:-1])
           
        elif s.endswith('k'):
            return float(s[:-1]) / 1024

        elif s == '1,000+':
            return 1.0
        
        else:
            return np.nan
    except:
        return np.nan
```

```python
df3['Size'] = df3['Size'].apply(fix_Size)
df3
```

```python
df3.info()
```

# **FIX PRICE**

```python
for p in df3['Price'].unique():
  print(p)
```

```python
def fix_Price(p):
  try:
    if p.startswith('$'):
      return float(p[1:])
    else:
        return 0.0
  except:
          return np.nan
```

```python
df3['Price'] = df3['Price'].apply(fix_Price)
df3
```
