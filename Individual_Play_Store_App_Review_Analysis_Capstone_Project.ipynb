#**Removing 90% empty columns**

#**DATA MANIPULATION**

```python
def assign_median(series):
  return series.fillna(series.median)
```

```python
df.Rating= df['Rating'].transform(assign_median)
```

```python
df.isnull().sum()
```

```python
print(df['Type'].mode())
print(df['Current Ver'].mode())
print(df['Android Ver'].mode())
```

```python
df['Type'].fillna(str(df['Type'].mode().values[0]), inplace = True)
df['Current Ver'].fillna(str(df['Current Ver'].mode().values[0]), inplace = True)
df['Android Ver'].fillna(str(df['Android Ver'].mode().values[0]), inplace = True)
```

```python
df.isnull().sum()
```

#**FIXING NUMERICAL DATA**



```python
for review in df['Reviews'].unique():
  print(review)
```

```python
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
df
```

```python
df.info()
```

```python
for s in df['Size'].unique():
  print(s)
```

```python
def fix_Size(s):
    try:
        if s.endswith('M'):
             return float(s[:-1])
           
        elif s.endswith('k'):
            return float(s[:-1]) / 1024

        elif s == '1,000+':
            return 1.0
        
        else:
            return np.nan
    except:
        return np.nan
```

```python
df['Size'] = df['Size'].apply(fix_Size)
df
```

```python
df.info()
```

# **FIX PRICE**

```python
for p in df['Price'].unique():
  print(p)
```

```python
def fix_Price(p):
  try:
    if p.startswith('$'):
      return float(p[1:])
    else:
        return 0.0
  except:
          return np.nan
```

```python
df['Price'] = df['Price'].apply(fix_Price)
df
```

```python
df.describe()
```

```python
for c in df['Category'].unique():
  print(c)
```

```python
df['Category'] = df['Category'].apply(lambda c: c.lower())

for c in df['Category'].unique():
  print(c)
```

```python
for i in df['Installs'].unique():
  print(i)
```

```python
for t in df['Type'].unique():
  print(t)
```

```python
for cr in df['Content Rating'].unique():
  print(cr)
```

```python
df.drop(['Genres' , 'Last Updated' , 'Current Ver'], axis=1, inplace=True)
df
```

```python
for v in df['Android Ver'].unique():
  print(v)
```

```python
def check_ver_4(version):
  try:
    if version.startswith('4'):
      return 'Yes'
    else:
        return ' No'
  except:
          return np.nan
    
```

```python
df['working_on_ver_4'] = df['Android Ver'].apply(check_ver_4)
df
```
